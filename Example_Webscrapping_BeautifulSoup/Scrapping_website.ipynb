```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
website_url = "http://pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/"
webpage = requests.get(website_url)
content = webpage.content
```

```python
soup_website = bs(content,"html.parser")
main_div = soup_website.find_all("div",{"class","propertyRow"})
```

```python
info_list = []
for each in main_div:
    temp_dict ={}
    temp_dict["Price"] = each.find("h4",{"class","propPrice"}).text.replace("\n","").replace(" ","")
    temp_dict["Address"] = each.find_all("span",{"class","propAddressCollapse"})[0].text
    temp_dict["Locality"] = each.find_all("span",{"class","propAddressCollapse"})[1].text
    try:
        temp_dict["Bed_Info"] = each.find("span",{"class","infoBed"}).find('b').text
    except:
        temp_dict["Bed_Info"] = 'None'
    try:
        temp_dict["Sqft_info"] = each.find("span",{"class","infoSqFt"}).find('b').text
    except:
        temp_dict["Sqft_info"] = 'None'
    try:
        temp_dict["ValueFullBath_info"] = each.find("span",{"class","infoValueFullBath"}).find('b').text
    except:
        temp_dict["ValueFullBath_info"] = 'None'
    try:
        temp_dict["ValueHalfBath_info"] = each.find("span",{"class","infoValueHalfBath"}).find('b').text
    except:
        temp_dict["ValueHalfBath_info"] = 'None'
        
    for column_group in each.find_all("div",{"class","columnGroup"}):
        #print(column_group)
        for feature_group,feature_name in zip(column_group.find_all("span",{"class","featureGroup"}),column_group.find_all("span",{"class","featureName"})):
            #print(feature_group.text,feature_name.text)
            if 'Lot Size' in feature_group.text:
                temp_dict["Lot Size"]=feature_name.text
    
    info_list.append(temp_dict)

#print(info_list)
```

```python
df = pd.DataFrame(info_list)
```
